```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
corona = pd.read_csv('2019_nCov_data.csv')
```

```python
corona.head()
```

```python
print(type(corona['Date'][0]))
```

```python
#converting the datetime from string to datetime object

corona['Date'] = pd.to_datetime(corona['Date'])
corona['Last Update'] = pd.to_datetime(corona['Last Update'])
```

```python
corona.head()
```

```python
corona.groupby(['Date']).max()
```

```python
corona['Date_Only'] = corona['Date'].dt.date
```

```python
#getting the total number of confirmed cases, deaths and reoveries
corona.groupby('Date_Only').sum()
```

```python
#get hour from the datetime column
corona['Hour'] = corona['Date'].dt.hour
```

```python
corona.head()
```

```python
corona.groupby('Hour').describe()
```

```python
corona['Date_Only'] = corona['Date_Only'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
#show the time series of the spread of the virus
days = [day for day, df in (corona.groupby('Date_Only'))]


plt.plot(days, corona.groupby(['Date_Only'])['Confirmed'].sum())
plt.xticks(days,rotation = 90)
plt.grid()
plt.show()
```

### The corona virus has been on the increase since it startes showing little signs of decline

```python
corona.groupby('Country')['Confirmed'].sum()
```

```python
#get the country associated with the confirmed cases
country = [country for country, df in corona.groupby('Country')]

plt.bar(country, corona.groupby(['Country']).sum()['Confirmed'])
plt.xticks(rotation = 90)
plt.yscale("log")
plt.show()
```

```python
country = [country for country, df in corona.groupby('Country')]

plt.bar(country, corona.groupby(['Country']).sum()['Deaths'])
plt.xticks(rotation = 90)
plt.yscale("log")
plt.show()
```

### Which country has the highest recovered Rate? Apparently China

```python
country = [country for country, df in corona.groupby('Country')]

plt.bar(country, corona.groupby(['Country']).sum()['Recovered'])
plt.xticks(rotation = 90)
plt.yscale("log")
plt.show()
```

```python
corona.groupby(['Province/State','Country']).sum()
```

```python
country = [country for country, df in corona.groupby('Province/State')]
plt.figure(figsize=(11,4))
plt.bar(country, corona.groupby(['Province/State']).sum()['Confirmed'], )
plt.xticks(rotation = 90)

plt.yscale("log")
plt.show()
```

### The hubei province in China clearly has the highest cases of the corona virus with other provinces following closely behind it
